# Juntando os dataframes

Nesse notebook será juntados os dois dataframes para servir de case para a palestra do sofwtare livre

```python
import pandas as pd
import numpy as np
```

```python
treino = pd.read_csv('Train.csv')
teste = pd.read_csv('Test.csv')
```

```python
treino.head()
```

Removendo as colunas que não vão ser necessárias para a anlise do treino e teste

```python
treino.drop(['ID','Var_1','Segmentation'],axis=1,
            inplace=True)
treino.head()
```

```python
teste.head()
```

```python
teste.drop(['ID','Var_1'],axis=1,inplace=True)
teste.head()
```

```python
seg_cliente = pd.concat([treino,teste])
seg_cliente.head()
```

```python
seg_cliente.info()
```

```python
seg_cliente.shape
```

Preferi apagar todas as linhas com valores livros mais por uma questão didática da apresentação, mas normalmente teria que fazer uma análise antes de fazer uma ação acerca dos valores nulos.

```python
seg_cliente.dropna(inplace=True)
seg_cliente.info()
```

Remoção de dados duplicados

```python
seg_cliente.drop_duplicates(inplace=True)
seg_cliente.info()
```

```python
seg_cliente.to_csv('Seg_clientes.csv')
```
